### 質的変数データの内容サマリー

```python
# 質的変数データの内容サマリー

# サマリー出力関数
def tmp_func_data_summary(target_col_name,output_limit):
    #tmp_df=pd.DataFrame(df_ksp.groupby(target_col_name).size().sort_index(),columns=["count"])
    tmp_df=df_ksp.groupby(target_col_name).size().sort_index().to_frame("count")
    record_num=len(tmp_df)
    output_num=record_num if record_num<=output_limit else output_limit
    print("================================================================================")
    print("===  項目名　：{}".format(target_col_name))
    print("===  表示形式：パターン(件数)")
    print("===  区切　　：／ ")
    print("===  表示数　：{}パターン中、{}パターン表示".format(record_num,output_num))
    print("===  以下、一覧")
    print("")
    print(' ／ '.join(map(lambda x: "{}({}件)".format(x,tmp_df.at[x,"count"]), tmp_df.head(output_limit).index)))
    if record_num>output_limit:
        print("上記、{}パターン中{}パターンまで表示。以降は省略".format(record_num,output_num)) 
    print("")
    
# 各変数の出力を実行
tmp_func_data_summary("state",100)
tmp_func_data_summary("main_category",100)
tmp_func_data_summary("category",100)
tmp_func_data_summary("country",100)
tmp_func_data_summary("currency",100)
tmp_func_data_summary("name",100)
```

###### note
* state "canceled","live","suspended","undefined"は、下記の理由により目的に適さないので、該当レコードを削除する。
 * canceled: 途中でキャンセルしており、案件の成功/失敗の結論が出ず終了している。
 * live: データ取得時点で進行中であり、案件の成功/失敗の結論が出ていない。
 * suspended: 途中で一時停止しており、案件の成功/失敗の結論が出ていない。
* 今回の目的に合わない項目がある。
 * pledged, usd_pledged, usd_pledged_real
 * backers
 * ID (ただし、調査段階では、データのIDとして使う可能性がある）
* 項目名に空白が入っている”usd pledged”はdf_ksp.usd pledgedと指定できないので扱いにくい。

## 分析方針の設定
* 教師データには成功/失敗のラベルが必要なので、成功/失敗（stateがsuccessfulとfailed）以外のデータは使用しない。削除してから分析する。
* 学習後の運用は、終了前の案件に対して成功/失敗を推定することになる。そのため、終了しなければ確定しない項目は説明変数に入れられない。具体的な項目としては、backers,pledged,usd_pledged,usd_pledged_realを説明変数に入れられない。
* 項目名に空白が入っている”usd pledged”は”usd_pledged”に変更して扱いやすくしてから分析する。
* IDには、今回の目的に使える情報は内在していないので、説明変数に入れられない。分析作業においてはデータのID機能として利用できる。

## 各項目の欠損状況の把握

```python
# 欠損調査 nan

print("---各項目におけるnan値の件数---")
display(df_ksp.isna().sum())

# 欠損調査 数値項目の値0
# 意味のある0かデータが欠損していることを示す0かを判断する必要あり
# pledged,usd_pledged_real,backersは説明変数として使えないから深入りしない。
print("---各項目における0値の件数---")
print("goal              ==0の件数 =",len(df_ksp[df_ksp.goal            ==0]))
print("usd_goal_real     ==0の件数 =",len(df_ksp[df_ksp.usd_goal_real   ==0]))
print("pledged           ==0の件数 =",len(df_ksp[df_ksp.pledged         ==0]))
print("backers           ==0の件数 =",len(df_ksp[df_ksp.backers         ==0]))
print("usd_pledged_real  ==0の件数 =",len(df_ksp[df_ksp.usd_pledged_real==0]))
```

###### note
* nameのnan値は3件と少ない。この項目から何らかの情報を抽出する方針になった場合でも、この3件はこのまま扱って良さそう。
* usd_pledgedのnan値は210件。何らかの値を補填すべきか検討する必要がある。→説明変数として使えない項目なので無視
* pledged,backers の0値は、0としての意味がある値なので、このまま扱う。→説明変数として使えない項目なので無視
* usd_pledged_real の0値は、何らかの補填を行うべきか検討する必要がある。→説明変数として使えない項目なので無視

### launched, deadlineの分析

```python
# 日付項目 launched,deadlineと成功/失敗の関係

tmp_date=df_ksp.copy()

# 8列を追加
# launched_dt　        datetime型の開始日
# deadline_dt　        datetime型の終了日
# funding_year         開催年（開始日から抽出）
# funding_start_month　開始月（開始日から抽出）
# funding_end_month　  終了月（終了日から抽出）
# funding_period_days　開催期間（日単位）（開始日と終了日の差、単位は日）
# funding_period_weeks 開催期間（週単位）（開始日と終了日の差、単位は週）
# start_end_of_month   開始終了月（開始月と終了月をハイフンで挟んだ文字列）
tmp_date["launched_dt"]            =pd.to_datetime(tmp_date.launched)
tmp_date["deadline_dt"]            =pd.to_datetime(tmp_date.deadline)
tmp_date["funding_year"]           =tmp_date.launched_dt.dt.year
tmp_date["funding_start_month"]    =tmp_date.launched_dt.dt.month
tmp_date["funding_end_month"]      =tmp_date.deadline_dt.dt.month
tmp_date["funding_period_of_days"] =(tmp_date.deadline_dt-tmp_date.launched_dt).dt.days
tmp_date["funding_period_of_weeks"]=(tmp_date.funding_period_of_days/7).astype(int)
tmp_date["start_end_of_month"]     =tmp_date.funding_start_month.astype(str).str.zfill(2)+"-"+tmp_date.funding_end_month.astype(str).str.zfill(2)



fig, ax = plt.subplots(3,1,figsize=(15,10))


# 開催期間（日単位）別の成功/失敗の色分けグラフ
sns.countplot(data=tmp_date,x="funding_period_of_days",hue="state",ax=ax[0])
ax[0].tick_params(axis='x', labelsize=8 )
ax[0].set_xlabel("period of days",fontsize=13)
ax[0].set_title("開催期間（日単位）と成否の関係",fontsize=20)
ax[0].legend(loc="best",fontsize=13)

# 開催期間（週単位）別の成功/失敗の色分けグラフ
sns.countplot(data=tmp_date,x="funding_period_of_weeks",hue="state",ax=ax[1])
ax[1].tick_params(axis='x', labelsize=8 )
ax[1].set_xlabel("period of weeks",fontsize=13)
ax[1].set_title("開催期間（週単位）と成否の関係",fontsize=20)
ax[1].legend(loc="best",fontsize=13)


# 開始終了月別の成功/失敗の色分けグラフ
sns.countplot(data=tmp_date,x="start_end_of_month",order=np.sort(tmp_date.start_end_of_month.unique()),hue="state",ax=ax[2])
ax[2].tick_params(axis='x', labelsize=6 )
ax[2].set_xlabel("[month of launched] - [month of deadline]",fontsize=13)
ax[2].set_title("開始月・終了月の組み合わせと成否の関係",fontsize=20)
ax[2].legend(loc="best",fontsize=13)


plt.tight_layout()
plt.show()


# ---サマリー表 tmp_date_summary の作成　開始-------
# tmp_date_summary の構造
#   インデックス１　開始月
#   インデックス２　開催期間（週単位）
#   成功件数
#   失敗件数
#   成功率

tmp_date_all     =tmp_date.groupby(["funding_start_month","funding_period_of_weeks"]                              ).size().to_frame("all_")
tmp_date_success =tmp_date[tmp_date.state=="successful"].groupby(["funding_start_month","funding_period_of_weeks"]).size().to_frame("success")
tmp_date_failed  =tmp_date[tmp_date.state=="failed"].groupby(["funding_start_month","funding_period_of_weeks"]    ).size().to_frame("failed")

tmp_date_summary=pd.merge(tmp_date_all,tmp_date_success,left_index=True,right_index=True)
tmp_date_summary=pd.merge(tmp_date_summary,tmp_date_failed,left_index=True,right_index=True)
tmp_date_summary["success_ratio"]= tmp_date_summary.success / tmp_date_summary.all_

del tmp_date_all
del tmp_date_success
del tmp_date_failed


# ---サマリー表 tmp_date_summary の作成　終了-------


# 3Dグラフ用メッシュグリッドデータ作成
tmp_X,tmp_Y=np.meshgrid(tmp_date_summary.index.get_level_values('funding_start_month').unique(),tmp_date_summary.index.get_level_values('funding_period_of_weeks').unique())
tmp_ufunc = np.frompyfunc((lambda x,y: tmp_date_summary.at[(x,y),"success_ratio"] if ((x,y) in tmp_date_summary.index) else 0),2,1)
tmp_Z=tmp_ufunc(tmp_X,tmp_Y)

# 開始月、開催期間（週単位）の２軸に対する成功率の３Dグラフ
fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(111, projection='3d')
aaa=ax.plot_surface(tmp_X,tmp_Y,tmp_Z, cmap='bwr', linewidth=0.1)
ax.contour(tmp_X,tmp_Y,tmp_Z, cmap='bwr', levels=10, offset = 0)
fig.colorbar(aaa, ax=ax)
ax.set_xlabel("the start month.")
ax.set_ylabel("the period of weeks.")
ax.set_zlabel("the ratio of success")
ax.set_title("開始月、開催期間（週単位）と成功率の関係",fontsize=20)
plt.show()

del tmp_X,tmp_Y,tmp_Z,tmp_date,tmp_date_summary,tmp_ufunc,aaa,fig,ax
```

## データのロードと加工

```python
# 元データCSVをロード
train_ksp=pd.read_csv("./in_data/ks-projects-201801.csv")

# canceled,live,suspended,undefinedのレコードを削除
train_ksp.drop(train_ksp[train_ksp.state.isin(["canceled","live","suspended","undefined"])].index,inplace=True)

# 開始月 funding_start_month と 開催期間（週単位）funding_period_of_weeks を生成
train_ksp["funding_start_month"]    =pd.to_datetime(train_ksp.launched).dt.month
train_ksp["funding_period_of_weeks"]=(
    (
        pd.to_datetime(train_ksp.deadline)-pd.to_datetime(train_ksp.launched)
    ).dt.days/7
).astype(int)

# nameの文字数 name_char_count を生成
train_ksp["name_char_count"]        =train_ksp.name.str.strip().str.len()
train_ksp.name_char_count.fillna(0,inplace=True)
train_ksp["name_char_count"]        =train_ksp["name_char_count"].astype(int)

# nameの単語数 name_word_count を生成
def tmp_func_word_count(data):
    counter=0
    for aaa in re.split('[,./\s]',data.strip()):
        counter+=1
    return counter
tmp_universal_func_word_count=np.frompyfunc(tmp_func_word_count,1,1)
train_ksp["name_word_count"]=tmp_universal_func_word_count(train_ksp.name.astype(str))
train_ksp["name_word_count"].fillna(0,inplace=True)
train_ksp["name_word_count"]=train_ksp["name_word_count"].astype(int)


# 使用しない項目の削除
train_ksp.drop(columns=["ID","goal","pledged","backers","usd pledged","usd_pledged_real","launched","deadline","name"],inplace=True)

display(train_ksp.info())
display(train_ksp.head())
```

## 量的変数の標準化

```python
tmp_target_col_list=["usd_goal_real","funding_start_month","funding_period_of_weeks","name_char_count","name_word_count"]
tmp_target=train_ksp[tmp_target_col_list]

# 標準化用オブジェクトの作成
scaler = StandardScaler() 
# 標準化のデータ解析実施。標準化の対象列、並び順も含めて間違えないように注意すること。
scaler.fit(tmp_target.astype(float))
# 標準化のデータ編集実施。対象列、並び順も含めて間違えないように注意すること。
tmp_target_scaled=scaler.transform(tmp_target.astype(float))


print("---標準化前のデータ　サンプル表示---------------------")
display(train_ksp.head())

train_ksp_scaled=train_ksp.copy()
train_ksp_scaled[tmp_target_col_list]=tmp_target_scaled

print("---標準化後のデータ　サンプル表示---------------------")
display(train_ksp_scaled.head())

print("---標準化解析結果　StandardScaler解析後パラメータ(項目名リスト、平均値リスト、分散リスト)-----")
print(tmp_target_col_list)
print(scaler.mean_)
print(scaler.var_)


print("---標準化前の各項目の統計情報---------------------")
for tmp_aaa in tmp_target_col_list:
    print("{} mean={:.5f}, min={:.5f}, max={:.5f}, std={:.5f}".format(
        tmp_aaa,
        train_ksp[tmp_aaa].mean(),
        train_ksp[tmp_aaa].min(),
        train_ksp[tmp_aaa].max(),
        train_ksp[tmp_aaa].std(),
    ))

    
print("---標準化後の各項目の統計情報---------------------") 
for tmp_aaa in tmp_target_col_list:
    print("{} mean={:.5f}, min={:.5f}, max={:.5f}, std={:.5f}".format(
        tmp_aaa,
        train_ksp_scaled[tmp_aaa].mean(),
        train_ksp_scaled[tmp_aaa].min(),
        train_ksp_scaled[tmp_aaa].max(),
        train_ksp_scaled[tmp_aaa].std(),
    ))

del tmp_target_col_list,tmp_target,tmp_target_scaled,tmp_aaa

```

## 質的変数データのダミー変数化

* 質的変数データをダミー変数化（one-hot encoding）する。

|ダミー化対象の変数|ダミー変数|説明|
|---|---|---|
|state|state_succsessful|successfulを1,failedを0で表現。|
|main_category|main_category_\[値\]|値ごとにダミー変数を用意。該当を1、非該当を0で表現。<br>値の種類だけダミー変数を用意する。|
|category|category_\[値\]|値ごとにダミー変数を用意。該当を1、非該当を0で表現。<br>値の種類だけダミー変数を用意する。|
|currency|currency_\[値\]|値ごとにダミー変数を用意。該当を1、非該当を0で表現。<br>値の種類だけダミー変数を用意する。|
|country|country_\[値\]|値ごとにダミー変数を用意。該当を1、非該当を0で表現。<br>値の種類だけダミー変数を用意する。|




```python
# category_encodersを使用する。
# fit()とtransform()を分けることができる点で、get_dummies()より良い。
# （get_dummies()は学習でfit()して予測対象データにtransform()できないので使えない。）
# sklearnのOneHotEncoderはDataFrameを扱えず、列名にエンコード対象カラムとラベル名を付けるのが、ミスる可能性を考えると不満。
# category_encodersなら、DataFrameを扱えて、エンコード対象カラムとラベルがエンコード後カラムの名前になるのでわかりやすい。



pd.set_option('display.max_columns', 10000)
pd.set_option('display.max_info_rows', 10000)
pd.set_option('display.max_info_columns', 10000)

print("---ダミー変数化する前の教師データ　サンプル表示--------------")
display(train_ksp_scaled.head())


# ダミー変数化するべき列は、数値型（int,float）以外の項目
train_ksp_scaled.info()
# category,main_category,currency,state,countryの５項目だと確認した。


# 対象列リスト
one_hot_target_col_list = ["category","main_category","currency","state","country"]

# OneHotEncoderオブジェクトを作成。
# handle_unknownオプションは、fit時に登場しなかったカテゴリをtransform時にデータ内に発見した場合にどう対処するか。
# 'ignore'なら無視する。'error'ならtransformをエラーで中止する。
one_hot_encoder = ce.OneHotEncoder(cols=one_hot_target_col_list,handle_unknown='ignore')
# 教師データを解析して、エンコーディング内容を決める。
one_hot_encoder.fit(train_ksp_scaled)
# 教師データに対して、エンコーディングを実施する。
train_ksp_scaled_dummied=one_hot_encoder.transform(train_ksp_scaled)

# エンコーディングの対応表を取得してDataFrame化
# ついでに、自動生成された列名と、変更してわかりやすくしたい新列名も生成する。
encode_map_list = []
for col_info in one_hot_encoder.category_mapping:
    for (k,v) in col_info['mapping'].items():
        encode_map_list.append([col_info['col'], k, v,col_info['col']+"_"+str(v),col_info['col']+"_"+str(k)])

encode_map_list= pd.DataFrame(encode_map_list, columns=['column', 'label', 'ord_num','encoded_column_generated','encoded_column_interpreted'])

print("---ダミー変数化（one-hot encoding）の変換内容　サンプル表示---------------------")
display(encode_map_list.head())


# 自動生成された列名をわかりやすい列名に変更
for row in encode_map_list.itertuples():
    col_name_old=row.encoded_column_generated
    col_name_new=row.encoded_column_interpreted
    train_ksp_scaled_dummied.rename(columns={col_name_old:col_name_new},inplace=True)

# 目的変数であるstateのsuccessful,failedだけはひと手間加える必要がある。
# 目的変数stateはダミー変数化後も１列にする必要がある。
# state_successfull 列の0,1だけでfailedも表現できるので、state_failed列をドロップするだけで良い。
train_ksp_scaled_dummied.drop(columns="state_failed",inplace=True)

print("---ダミー変数化＆列名変更後の教師データ　サンプル表示--------------")
display(train_ksp_scaled_dummied.head())

del col_info,col_name_old,col_name_new,k,row,v
```

### 予測結果の評価

```python
# 対数尤度を算出（方法１）
# log_lossは”負の”対数尤度を計算する関数であるため、マイナスを付けて本来の対数尤度に戻す。
# normalizeオプションがTrueのままでは、合計の対数尤度ではなく平均の対数尤度が計算されるので注意。合計の対数尤度を求めたいのでFalseとする。
log_likelihood1 = - log_loss(train_Y_1d, y_est_proba,normalize=False)
print("対数尤度（log_loss関数出力版）：{:.3f}".format(log_likelihood1))

# 対数尤度を算出（方法２）
# ライブラリを使わず、自分で計算
log_likelihood2 = 0
epsilon = 1e-15
for i in range(len(train_Y_1d)):
    log_likelihood2 += train_Y_1d[i]*np.log(y_est_proba[i,1]+epsilon)+(1-train_Y_1d[i])*np.log(y_est_proba[i,0]+epsilon)

print("対数尤度（自力計算出力版）　　：{:.3f}".format(log_likelihood2))



# ライブラリを使用して評価指標を算出
accuracy = accuracy_score(train_Y_1d, y_est)
precision, recall, f1_score, _ = precision_recall_fscore_support(train_Y_1d, y_est)


# 評価指標の表示
print("---ライブラリを使用して算出した評価指標-----------")
print("正答率(Accuracy)  = {:.3f}%".format(100 * accuracy    ))
print("適合率(Precision) = {:.3f}%".format(100 * precision[1]))
print("再現率(Recall)    = {:.3f}%".format(100 * recall[1]   ))
print("F1値(F1-score)    = {:.3f}%".format(100 * f1_score[1] ))


# 混同行列を表示
display(
    pd.DataFrame(
        confusion_matrix(train_Y_1d,y_est),
        index=["真（failed）","真（successful）"],
        columns=["推定（failed）","推定（successful）"])
       )

# 混同行列から、手計算で評価指標を算出
print("---混同行列から手計算で算出した評価指標-----------")
print("正答率accuracy  ={:.3f}".format(  (157444+65673)/(157444+40275+68283+65673)))
print("適合率precision ={:.3f}".format(          65673 /(       40275+      65673)))
print("再現率recall    ={:.3f}".format(          65673 /(             68283+65673)))
```

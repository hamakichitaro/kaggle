## 共通処理の関数定義

```python
# 複数セルで何度も実行する処理のうち、
# 共通化することで全体の見通しが良くなる処理について、本セルで関数定義する。
# 単一セル内でのみ使用する関数定義は、そのセル内で定義することとする。


def my_get_state_summary(tmp_df, group_by_keys_array):
    """
    kickstarterプロジェクトのデータ（DataFrame）を参照して、成功率サマリー表（DataFrame）を作成して返す。
    成功率サマリー表とは、データ（tmp_df）に対して、group_by_keys_arrayでグルーピングして、state列の値を集計したものである。
    集計内容はグルーピングキーをインデックスとして、成功件数、失敗件数、全件数、成功率の４項目を集計している。
    Parameters
    ----------
    tmp_df : pandas.DataFrame
        kickstarterプロジェクトのデータ構造。state列を持ち、state列の値は'successful'と'failed'の２値。None不許可
    group_by_keys_array : list of labels
        tmp_dfに存在する列名を要素としたリスト。この列名でグルーピングして集計される。
        複数要素がある場合は要素順がグルーピング順になる。
    Returns
    ----------
    tmp_df_summary : pandas.DataFrame
        成功率サマリー表
        グルーピングキーをインデックスとして持ち、下記４列を持つDataFrame。
            success : 'successful'の件数カウント
            failed : 'failed'の件数カウント
            all_ : 全件数カウント（'successful'の件数カウント＋'failed'の件数カウント）、
            success_ratio : 'success'の件数カウント/'all_'の件数カウント
        グルーピングキーが複数ある場合はマルチインデックスになる。
    Examples
    ----------
    success_summary_table = get_success_summary_data(df_kickstarter,["main_category","category"])
    """
    tmp_df_all     =tmp_df.groupby(group_by_keys_array).size().to_frame("all_")
    tmp_df_success =tmp_df[tmp_df.state=="successful"].groupby(group_by_keys_array).size().to_frame("success")
    tmp_df_failed  =tmp_df[tmp_df.state=="failed"].groupby(group_by_keys_array).size().to_frame("failed")
    tmp_df_summary=pd.merge(tmp_df_all,tmp_df_success,left_index=True,right_index=True)
    tmp_df_summary=pd.merge(tmp_df_summary,tmp_df_failed,left_index=True,right_index=True)
    tmp_df_summary["success_ratio"]= tmp_df_summary.success / tmp_df_summary.all_
    del tmp_df_all
    del tmp_df_success
    del tmp_df_failed
    return tmp_df_summary



def my_stratified_sampling(source_df,rate,stratify,random_state):
    """
    割合を指定して、ランダムにサンプル抽出する。
    stratifyの列で層別に実施することで、stratifyの分布状況を維持する。
    
    Parameters
    ----------
    source_df : pandas.DataFrame
        サンプリング対象となる元データ
    rate : float
        サンプリングする割合
    stratify : str
        層別のキーとなる列名
    random_state : int
        ランダム抽出のシード

    Returns
    ----------
    pandas.DataFrame
        source_dfから、stratifyで層別にrateの割合だけサンプル抽出したデータ
    
    Examples
    ----------
    cv_large_XY   =my_stratified_sampling(source_df=train_XY,rate=0.5,stratify="state_successful",random_state=1234)
    cv_medium_XY  =my_stratified_sampling(source_df=train_XY,rate=0.3,stratify="state_successful",random_state=1234)
    cv_small_XY   =my_stratified_sampling(source_df=train_XY,rate=0.1,stratify="state_successful",random_state=1234)
    """
    sample=source_df.groupby(stratify).apply(
        lambda x: x.sample(
            n=round( len(x) * rate ) ,
            axis=0,
            random_state=random_state,
            replace=False
        )
    )
    sample.reset_index(level=0,drop=True,inplace=True)
    return sample


def my_extract_zip_and_load_pickle(zip_file_path, extract_dest_dir_path, extract_target_file_name):
    """
    ZIPファイルを展開して、対象のpickleファイルをロードする。
    ロードが完了したら、展開したファイルは削除してこの関数を終える。
    ロードしたオブジェクトを関数の戻り値として返す。
    
    Parameters
    ----------
    
    zip_file_path : str
        ZIPファイルのパス
    extract_dest_dir_path : str
        展開先ディレクトリのパス
    extract_target_file_name : str
        ロード対象ファイル名
    Returns
    ----------
    object
        pickleファイルをロードしたオブジェクト
    Examples
    ----------
    new_object=my_extract_zip_and_load_pickle("./out_data/data_dict.zip","./out_data","data_dict.pkl")
    """
    
    with zipfile.ZipFile(zip_file_path) as f:
        # zipファイルを展開
        f.extractall(extract_dest_dir_path)

        # ロード
        tmp_object=pd.read_pickle(extract_dest_dir_path+"/"+extract_target_file_name)
        # zipファイルが残っていれば良いので、展開後ファイルは削除
        os.remove(extract_dest_dir_path+"/"+extract_target_file_name)
        
    return tmp_object
    
```

## nameの単語分析

nameはそのままでは使えないが、使用されている単語から、特徴を抽出できるかもしれない。<br>
使用単語毎にone-hot化することを想定して、単語リストを作成して検討する。<br>
どんな単語が抽出されたか、単語数などを確認して、特徴量化する価値があるか判断する。
<br><br>
###### (仮説)
案件名で使用されている単語が案件の魅力を左右して、案件成功率に影響する。<br>
複数の単語の繋がりで生まれる特徴については、今回は無視する。あくまで単語単体での特徴に限定して扱う。<br>
下記３条件を満たす単語の使用有無が、案件成功率に影響する。<br>
* 案件の成否に偏りがある単語は、案件成否への影響が大きい単語とみなせる。
* 使用頻度 N回 以上の単語、つまりN 案件以上で使用されている単語は、サンプルとして信頼性があるとみなせる。（Nは分析結果を見て決める）。
* 単語単体では案件成否に影響が無いと想定される単語は分析対象から除外する。


###### (特徴抽出手順)

* 詳細分析セクション（本セクション）での処理
 - 使用頻度N以上の単語、つまりN案件以上で使用されている単語を、サンプルとして信頼性があるとみなして抽出する。<br>同時に、単語毎に成功案件数と失敗案件数を集計して、成功率を求めてランキング化する。（使用単語リスト＿全単語版）
 - 単体では重要な意味を成さない単語のリスト（除外対象単語リスト）の該当を、使用単語リスト＿全単語版から除外する。
 - ランキングが高い単語と低い単語（成功率が高い単語と成功率が低い単語）を、特徴量化の対象単語としてリスト化する。（使用単語リスト＿完成版）

* 前処理セクションでの処理
 - 対象単語リストにある単語毎にone-hot化する。その案件で使用されていたら1を立てる。

##### 使用単語リスト＿全単語版　の作成

```python
# %%timeは並列処理でうまく動作しないので、time()を使う。
start = time.time()


# 作業に必要なname,stateだけのデータフレームを作成する。
tmp_name=df_ksp.loc[:,["name","state"]]

# 処理時間短縮のため、分割して並列処理する。各ワーカープロセスに任せるデータ件数を設定。
batch_size= 50000

# バッチ数の算出
workers_size=math.ceil(tmp_name.shape[0] / batch_size)

print("データ件数=",tmp_name.shape[0])
print("バッチサイズ=",batch_size)
print("バッチ数=",workers_size)

# データを各ワーカープロセスに振り分けるためにバッチ分けしてリスト化
tmp_name_batch_list=[]
for i in range(1,workers_size+1):
    from_idx =batch_size * (i-1)
    to_idx   =batch_size * i if tmp_name.shape[0] > (batch_size * i) else tmp_name.shape[0]
    tmp=tmp_name[from_idx:to_idx]
    print("バッチNo.={:},\t行番号範囲={:} - {:},\t件数={:}".format(i,tmp.head(1).index[0],tmp.tail(1).index[0],tmp.shape[0]))
    tmp_name_batch_list.append(tmp)

    
# 並列処理内容を定義
# 各ワーカープロセスでは、担当したデータの各案件名nameで使われている単語を調べた結果をデータフレームにして戻す。
def parallel_task_word_count(df):
    # 単語をインデックスにした、成功件数と失敗件数の２列のデータフレーム作成。wcはword countの略
    tmp_wc_rank_list=pd.DataFrame(index=[],columns=["successful","failed"])
    # データの各行で、nameを単語に分解して成功失敗を計数する。
    for row in df.itertuples():
        if (row.name is None) | (type(row.name) is float):
            continue
        
        # 区切り記号で単語に分割して、単語毎に繰り返す。
        # １案件内で単語が何回使用されているかは無視して、１回以上使用されていれば頻度１とする。
        # １案件内で複数回使用されていても１とカウントする処理は、案件ごとで一旦セットにまとめることで実現する。
        words_set=set()        
        for www in (aaa.lower() for aaa in re.split("[<>~#®•=—@\*\?\|\'\"\(\)\{\}\[\]\_\-\–\+;:/,.!\s]",row.name.strip()) if aaa !=''):
            words_set.add(www)
        
        # セットに登録された単語毎に、ランキング表に追加する。
        for www in words_set:
            # その単語の行がランキング表に存在しなければ、行を追加する。
            if www not in tmp_wc_rank_list.index.values:
                tmp_wc_rank_list.loc[www]=[0,0]
            # 既存行に+1する。
            tmp_wc_rank_list.loc[www,row.state]+=1
        
    return tmp_wc_rank_list


# ワーカープロセスを並列で生成して、並列処理させる。
# バッチ分けしたリストを引数にすることで、リストの各要素（バッチ）を各ワーカーに引き渡す。
# 各ワーカープロセスからの戻り値である使用頻度ランキング表を回収する。
with concurrent.futures.ProcessPoolExecutor(max_workers=workers_size) as executor:
    result_list=list(executor.map(parallel_task_word_count, tmp_name_batch_list))

    
# 各ワーカープロセスが生成した使用頻度ランキング表を統合して、全体としての使用頻度ランキング表を生成する。
integrated_wc_rank_list=pd.DataFrame(index=[],columns=["successful","failed"])
tmp_count=0
for aaa in result_list:
    tmp_count += 1
    print("---使用頻度ランキング表の統合 {}/{}.---".format(tmp_count,len(result_list)))
    for row in aaa.itertuples():
        if row.Index not in integrated_wc_rank_list.index.values:
            integrated_wc_rank_list.loc[row.Index]=[0,0]
        integrated_wc_rank_list.loc[row.Index,"successful"]+=row.successful
        integrated_wc_rank_list.loc[row.Index,"failed"]+=row.failed

# 統合版ランキング表で、all_、success_ratioの２列を追加（属性allと競合するので_を付けてall_にしている）
integrated_wc_rank_list["all_"]          =integrated_wc_rank_list.successful + integrated_wc_rank_list.failed
integrated_wc_rank_list["success_ratio"] =integrated_wc_rank_list.successful / integrated_wc_rank_list.all_
# 統合版ランキング表を、単語の使用頻度で降順ソート。第２キーは表の画面出力の視認性を上げるのが目的。
# 本当は、使用頻度all_と単語名（インデックス）の複合キーでソートしたいが、やり方がわからないので保留。
integrated_wc_rank_list.sort_values(["all_","success_ratio"], ascending=[False,False],inplace=True)

print()
print()
print("使用単語リスト＿全単語版のサンプル出力")
display(integrated_wc_rank_list.head(10))

stop = time.time()
print('%.3f seconds' % (stop - start))

del tmp_name
del batch_size,workers_size,tmp_name_batch_list,executor,result_list
del aaa, i, row, from_idx, to_idx,tmp,tmp_count
del start, stop
```

##### 使用単語リスト＿完成版の作成

本セルでの処理内容
* 除外対象リストの該当単語をリストから削除
* 使用頻度の低い単語をリストから削除
* 案件成否への影響が薄い単語をリストから削除

```python
start = time.time()

# ランキング表の再生成は時間がかかるので、
# 作業の便宜上、コピーを生成して作業を進める。（このセルだけやり直しがきく。）
new_integrated_wc_rank_list=integrated_wc_rank_list.copy()

# 除外対象リストexclude_wordsの単語をランキング表から削除する。
for bbb in exclude_words:
    if bbb in new_integrated_wc_rank_list.index.values:
        new_integrated_wc_rank_list.drop(index=bbb,inplace=True)


# 使用頻度の低い単語は、特徴のサンプルとして信頼度が低い。また、成功/失敗のどちらかに偏る単語しか、予測に影響を及ぼさないと言える。
# 閾値を設けて切り落とす。
# さらにsuccess_ratioが中途半端な単語は、案件成否に寄与しないと考えて切り落とすr。

# 閾値を変えて、使用単語リスト＿完成版を４パターン作成する。
# 　使用頻度の閾値100、success_ratioは>=0.7 or <=0.3
# 　使用頻度の閾値200、success_ratioは>=0.7 or <=0.3
# 　使用頻度の閾値500、success_ratioは>=0.7 or <=0.3
# 　使用頻度の閾値1000、success_ratioは>=0.7 or <=0.3
one_hot_target_word_list={}
for threshold in [100,200,500,1000]:
    one_hot_target_word_list[threshold]=new_integrated_wc_rank_list[
        (new_integrated_wc_rank_list.all_>=threshold) & 
        ((new_integrated_wc_rank_list.success_ratio>=0.7)|(new_integrated_wc_rank_list.success_ratio<=0.3))
    ].index.values


stop = time.time()
print('%.3f seconds' % (stop - start))

del bbb, new_integrated_wc_rank_list
del start,stop, threshold
```

## nameから特徴抽出（使用単語one-hot化）

```python
start=time.time()


# 作業対象列だけのデータフレームを作成
tmp_name=train_ksp.copy().name.to_frame()

# 対象リストをロード（閾値100のリストを使用）
in_filename="./out_data/one_hot_target_word_list_threshold100.txt"
with open(in_filename, mode="r") as f:
    one_hot_target_word_list=f.read() .split("\n")
print("読み込みファイル：",in_filename)
print("単語数　　　：",len(one_hot_target_word_list))
print("単語内容　　："," / ".join(one_hot_target_word_list))
print()


# リストにある単語について、one-hot化する。

# ダミー変数列を追加
for word in one_hot_target_word_list:
    tmp_name["name_wd_"+word]=0

# CPU数に合わせてバッチ分けするための、バッチサイズとバッチ数（ワーカー数）算出
import os
num_process=os.cpu_count()
batch_size= int(tmp_name.shape[0]/num_process)
workers_num=math.ceil(tmp_name.shape[0]/batch_size)

print("データ件数=",tmp_name.shape[0])
print("バッチサイズ=",batch_size)
print("バッチ数=",workers_num)

# データをワーカープロセスに振り分けるためにバッチ分けしてリスト化
batch_list = [tmp_name[i:((i+batch_size) if i+batch_size<tmp_name.shape[0] else tmp_name.shape[0])] for i in range(0, tmp_name.shape[0], batch_size)]



# 並列処理内容を定義
def parallel_task_word_one_hot_encode(df):
    # nameを単語に分解して、対象単語リストに存在する場合は該当列（ダミー変数）に1を立てる。
    for idx,row in df.iterrows(): # itertuples()では、カラム数が多くなるとnamedtupleではなくtupleになってしまう!!!!!!!!!
        if (row["name"] is None) | (type(row["name"]) is float):
            continue
        # nameで使用されている単語の種類を抽出。一旦セットに格納することで、種類数としてまとめる。
        words_set=set()
        for www in (aaa.lower() for aaa in re.split("[<>~#®•=—@\*\?\|\'\"\(\)\{\}\[\]\_\-\–\+;:/,.!\s]",row["name"].strip()) if aaa !=''):
            words_set.add(www)
        
        # セットに登録された単語毎に、対象単語リストと突合。ヒットしたらダミー変数に1を立てる。
        for www in words_set:
            if www in one_hot_target_word_list:
                df.loc[idx,"name_wd_"+www]=1
    return df


# ワーカープロセスを生成して、並列処理させる。
# バッチ分けしたリストを引数にすることで、リストの各要素（バッチ）を各ワーカーに引き渡す。
# 各ワーカープロセスの処理結果は、引数として与えたデータのバッチに反映されている。
with concurrent.futures.ProcessPoolExecutor(max_workers=workers_num) as executor:
    result_list=list(executor.map(parallel_task_word_one_hot_encode, batch_list))

# 処理結果が反映された各バッチを統合して、処理結果データフレームを生成する。
tmp_name=pd.concat(result_list)

print("---one-hot結果のサンプル表示--------------------")
display(tmp_name.head())
tmp_name.info()



# 生成したone-hot列をデータ本体に統合する。
train_ksp=pd.merge(train_ksp,tmp_name.drop(columns=["name"]),left_index=True,right_index=True)



stop=time.time()
print("%.3f seconds" % (stop - start))

del tmp_name, in_filename, f
del num_process,batch_size,workers_num,batch_list, result_list
del executor,one_hot_target_word_list, word
del start, stop
```

## 量的変数の多項式化

PolynomialFeaturesを利用して、量的変数を多項式化して説明変数に追加する。<br>
追加するのは、量的変数４項目（usd_goal_real,funding_period_of_weeks,name_char_count,name_word_count）を元に生成した４次までの多項式項とする。<br>
つまり、PolynomialFeaturesにオプション degree=4,interaction_only=False, include_biasFalse　で４項目を与えて得られた結果である。<br>
無意味な項もあるかもしれない（全項目が無意味かもしれない）が、最終的には特徴選択でふるいにかけられれば良いと考える。

```python
# 対象とする項目名のリスト化
poly_target_col_list=["usd_goal_real","funding_period_of_weeks","name_char_count","name_word_count"]


# 作業対象列だけのデータフレームを作成
poly_input =train_ksp[poly_target_col_list]

# 定数項なし、4次の多項式行列を作るインスタンスを構築する。
poly=PolynomialFeatures(degree=4,include_bias=False,interaction_only=False)

# インプットに対する多項式行列を得る。
poly_output =poly.fit_transform(poly_input)

# DataFrame化する。インデックスは元データの内容を引き継ぐ。多項式行列の4列目以降のみ用いる（左４項目は元データに既に存在するから。）
df_poly_output=pd.DataFrame(
    poly_output,
    index=poly_input.index,
    columns=poly.get_feature_names(input_features=poly_input.columns)
).drop(columns=poly_input.columns)

display(df_poly_output.head())
df_poly_output.info()

# インデックスをキーに、元データに連結する。
train_ksp=pd.merge(train_ksp,df_poly_output,left_index=True,right_index=True)


# 生成した列名をリストとして確保する。「量的変数の標準化」セクションで使用する。
poly_col_names=df_poly_output.columns.to_list()

del poly_target_col_list, poly, poly_input, poly_output, df_poly_output
```

## 量的変数の主成分分析による無相関化検討（本データに適用せず）

PCAを使用して、量的変数に対して無相関化を検討する。<br>
ただし、本データへの適用は行わないこととする。<br>
特徴選択で説明変数を減らすので、ここで次元削減する必要は無いから。また、特徴選択時に取捨判定する説明変数の解釈性を保ちたいから。

```python
start=time.time()
#対象列名リストの作成
pca_target_col_list=standard_scaler_target_col_list
print("PCA対象列数：",len(pca_target_col_list))

#対象データだけのデータフレーム化
pca_input=train_ksp[pca_target_col_list]

#主成分軸の次元数。
pca_n_components=len(pca_target_col_list) #次元削減はしない。
# PCAオブジェクトの作成
pca = PCA(n_components=pca_n_components)  

# PCA解析実施。同時にデータ編集実施。
pca_output=pca.fit_transform(pca_input)


# PCA結果のデータフレーム化。処理前のデータフレームのインデックスを引き継ぐ。
pca_output=pd.DataFrame(
    pca_output,
    index=pca_input.index,
    columns=["PC{}".format(x + 1) for x in range(pca_n_components)]
)


# 本データに反映する。←反映しないことにした!!!　コメントアウトしておく。
# train_ksp.drop(columns=pca_target_col_list,inplace=True)
# train_ksp=pd.merge(train_ksp,pca_output,left_index=True,right_index=True)


print("---無相関化解析結果　---------------------")
print("各主成分の寄与率=",pca.explained_variance_ratio_)
print("各主成分の平均　=",pca.mean_)
print("各主成分の固有ベクトル　=",pca.components_)
print()
print()

# データフレーム化してグラフ表示
pca_explained_variance_ratio=pd.DataFrame(pca.explained_variance_ratio_,columns=["explained_variance_ratio"],index=pca_output.columns)
pca_explained_variance_ratio.plot.barh(figsize=(20,15),title="各主成分軸の寄与率",legend=None)

end=time.time()
print("%f seconds." % (end-start))


del pca_explained_variance_ratio
del pca_input
del pca_n_components
del pca_output
del start,end
#del pca
```

##### note
寄与率を見ると、無相関化を試みた69列は、主成分20軸、つまり20列程度に集約できることが分かった。

##### 判断
とりあえず、無相関化は本データに適用しないこととする。<br>
特徴選択で変数を削減する方針とする。<br>
理由は、無相関化すると変数の解釈性がなくなり、特徴選択の段階で、どの特徴が重要なのか判断できなくなってしまうから。

## 特徴選択

「埋め込み法＋DecisionTreeClassifier」で特徴選択を実施する。<br>
グリッドサーチでDecisionTreeClassifierのハイパーパラメータを探索して、ベストパラメータを求める。<br>
そのベストパラメータを用いたDecisionTreeClassifierとSelectFromModelで特徴選択する。<br>
<br><br> 
##### 理由
精度が高いのはラッパー法だが、データ全件に対して処理すると処理時間が長くなってしまう。<br>
Day3では、埋め込み法＋DecisionTreeで全件に対して現実的な時間で処理できた。

```python
start=time.time()

# ----------------------------------------------------------
# データの準備
train_X=train_ksp.drop(columns="state_successful")
train_Y_1d=train_ksp["state_successful"].values.ravel()


# ----------------------------------------------------------
# 特徴選択に使う評価器モデルのハイパーパラメータ探索

# グリッドサーチするハイパーパラメータセット
parameters=[
    {'max_depth' : [100,1000,2000],'min_samples_leaf' : [100,1000,2000,3000],'min_samples_split' : [100,1000,2000,3000]}
]

# モデル構築
model = DecisionTreeClassifier(criterion="gini", random_state=1234)

# グリッドサーチを構築する。
clf = GridSearchCV(model,parameters, cv=5,n_jobs=-1,verbose=1)

# 学習する。
clf.fit(train_X,train_Y_1d)

# グリッドサーチが判定した最適なハイパーパラメータセットを出力する。
print("ベストパラメータとスコア：",clf.best_params_, clf.best_score_)


# ----------------------------------------------------------
# 特徴選択

# 評価器の構築
estimator = DecisionTreeClassifier(**clf.best_params_,criterion="gini", random_state=1234)

# モデルの情報を使って特徴選択を行う場合は、SelectFromModelを使う
# 今回は重要度が1e-3以下である特徴を削除する
# 重要度のしきい値はthresholdで指定する
sfm = SelectFromModel(estimator, threshold=1e-3)
sfm.fit(train_X,train_Y_1d)


print("---削除すべき説明変数---")
print(train_X.columns[~sfm.get_support()].values)

# 削除対象のリストを後で使うために残しておく。
feature_selection_result=train_X.columns[~sfm.get_support()].values

# 評価器で得た各特徴の重要度の値を棒グラフで表示
ft_importances=sfm.estimator_.feature_importances_
df_ft_importances=pd.DataFrame(ft_importances)
df_ft_importances.index=train_X.columns
df_ft_importances.plot.barh(figsize=(20,80),title="評価器で得た説明変数の重要度（不純度減少量合計）",legend=None)
plt.show()


end=time.time()
print("%f seconds." % (end-start))

del model,parameters,clf,
del estimator, ft_importances, df_ft_importances
del train_X,train_Y_1d
```

```python
# 本データに反映
train_ksp.drop(columns=feature_selection_result,inplace=True)
print("特徴選択で残った変数")
train_ksp.info(max_cols=len(train_ksp.columns))
```

### 探索アルゴリズム評価４　ベイズ最適化

パラメータ３種ともに「2〜2000の整数からランダムサンプリング」というルールを与えて、あとはベイズ最適化のアルゴリズムに任せる。アルゴリズム内での自動探索500回試行するオペレーションを１回実行する。

```python
start=time.time()

# ----------------------------------------------------------
# データの準備
data_dict=my_extract_zip_and_load_pickle("./out_data/data_dict.zip", "./out_data", "data_dict.pkl")

cv_X , cv_y  =data_dict["cv_medium"]

train_X, train_y   =data_dict["all_train"]
train_y = train_y.values.ravel()

test_X , test_y   =data_dict["all_test"]
test_y = test_y.values.ravel()




# ----------------------------------------------------------
# ベイズ最適化
def objective(trial):
    
    #ハイパーパラメータ
    parameters= {
        'max_depth'         : trial.suggest_int("max_depth",         2, 2000,1),
        'min_samples_leaf'  : trial.suggest_int("min_samples_leaf",  2, 2000,1),
        'min_samples_split' : trial.suggest_int("min_samples_split", 2, 2000,1),
    }
    
    
    #交差検証
    cv_counter=0
    sum_score=0
    for train_idx, val_idx in KFold(n_splits=5, shuffle=True, random_state=1234).split(cv_X,cv_y):
        cv_counter+=1
        # 訓練用データ
        cv_X_tr = cv_X.iloc[train_idx,:]
        cv_y_tr = cv_y.iloc[train_idx,:].values.ravel()
        # 検証用データ
        cv_X_val = cv_X.iloc[val_idx,:]
        cv_y_val = cv_y.iloc[val_idx,:].values.ravel()
        
        #評価器
        model =DecisionTreeClassifier(**parameters,criterion="gini", random_state=1234)
        model.fit(cv_X_tr,cv_y_tr)
        sum_score+=model.score(cv_X_val,cv_y_val)
    
    return sum_score/cv_counter #平均スコア


study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)
print("ベストパラメータ：",study.best_params)
print("交差検証スコア(accuracy)：",study.best_value) 



# ----------------------------------------------------------
# ベイズ最適化で決定した最適なハイパーパラメータを用いてモデルを構築して、最終テストを行う。

best_model =DecisionTreeClassifier(**study.best_params,criterion="gini", random_state=1234)
best_model.fit(train_X,train_y)
print("最終スコア(accuracy)=",best_model.score(test_X, test_y)) #accuracyのスコアを表示している。

end=time.time()
print("%f seconds." % (end-start))

del data_dict
del cv_X,cv_y,train_X,train_y,test_X,test_y
del start,end
```

### 探索アルゴリズム評価サマリー

方法論としてはベイズ最適化がとても良さそうだが、手間暇をかけても良いならグリッドサーチで確実に絞りんで行く方法も捨てられない。<br>
グリッドサーチで絞り込みながら、絞り込んだ限定範囲をランダムサーチ（確率分布）やベイズ最適化で探索する方法が良いのかもしれない。<br>
もっと知見と経験を増やして、目的や状況に応じて使い分けられるようになりたい。


|　|グリッドサーチ|ランダムサーチ（パターンリスト）|ランダムサーチ（確率分布）|ベイズ最適化|

### 分類モデル評価７　KNeighbors

```python
start=time.time()

# ----------------------------------------------------------
# データの準備
data_dict=my_extract_zip_and_load_pickle("./out_data/data_dict.zip", "./out_data", "data_dict.pkl")

cv_X , cv_y  =data_dict["cv_small"]

train_X, train_y   =data_dict["all_train"]
train_y = train_y.values.ravel()

test_X , test_y   =data_dict["all_test"]
test_y = test_y.values.ravel()


# ----------------------------------------------------------
# ベイズ最適化
def objective(trial):
    
    #ハイパーパラメータ
    parameters= {
        'n_neighbors'  : trial.suggest_int("n_neighbors", 5, 100,5),
        'weights'      : trial.suggest_categorical("weights", ["uniform","distance"]),
        'leaf_size'    : trial.suggest_int("leaf_size", 30, 90,30),
    }
    
    
    #交差検証
    cv_counter=0
    sum_score=0
    for train_idx, val_idx in KFold(n_splits=5, shuffle=True, random_state=1234).split(cv_X,cv_y):
        cv_counter+=1
        # 訓練用データ
        cv_X_tr = cv_X.iloc[train_idx,:]
        cv_y_tr = cv_y.iloc[train_idx,:].values.ravel()
        # 検証用データ
        cv_X_val = cv_X.iloc[val_idx,:]
        cv_y_val = cv_y.iloc[val_idx,:].values.ravel()

        #評価器
        model =KNeighborsClassifier(algorithm="auto",**parameters)
        model.fit(cv_X_tr,cv_y_tr)
        sum_score+=model.score(cv_X_val,cv_y_val)
    
    return sum_score/cv_counter #平均スコア


study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100,n_jobs=-1)
print("ベストパラメータ：",study.best_params)
print("交差検証スコア(accuracy)：",study.best_value) 


# ----------------------------------------------------------
# ベイズ最適化で決定した最適なハイパーパラメータを用いてモデルを構築して、最終テストを行う。

best_model =KNeighborsClassifier(algorithm="auto",**study.best_params)
best_model.fit(train_X,train_y)
print("最終スコア(accuracy)=",best_model.score(test_X, test_y)) #accuracyのスコアを表示している。

end=time.time()
print("%f seconds." % (end-start))

del data_dict
del cv_X,cv_y,train_X,train_y,test_X,test_y
del start,end
```

* nbextensionsにて、拡張機能 Table of Contents (2) を有効化<br>（見出し要素に番号付与、文書頭に目次リンク生成、jupyterエディタ上に目次タブ表示）<br><br>

## 共通処理の関数定義

```python
# 複数セルで何度も実行する処理のうち、
# 共通化することで全体の見通しが良くなる処理について、本セルで関数定義する。
# 単一セル内でのみ使用する関数定義は、そのセル内で定義することとする。


# 成功率サマリー表 作成
# 引数で指定したデータフレームに対して、グルーピング項目で指定した項目毎に成功件数、失敗件数、成功率を集計し、結果をDataFrameとして出力する。
# 引数)
# 　tmp_df    元データのDataFrame。引数group_by_arrayで指定したグルーピング項目を列として持っている必要あり。
#             またstate列を持っており、state列には"successful","failed"が設定されている必要あり。
# 　group_by_keys_array  グルーピング項目の配列
# 例）
# 　success_summary_data = get_success_summary_data(df_ksp,["funding_start_month","funding_period_of_weeks"])

def my_get_state_summary(tmp_df, group_by_keys_array):
    tmp_df_all     =tmp_df.groupby(group_by_keys_array).size().to_frame("all_")
    tmp_df_success =tmp_df[tmp_df.state=="successful"].groupby(group_by_keys_array).size().to_frame("success")
    tmp_df_failed  =tmp_df[tmp_df.state=="failed"].groupby(group_by_keys_array).size().to_frame("failed")
    tmp_df_summary=pd.merge(tmp_df_all,tmp_df_success,left_index=True,right_index=True)
    tmp_df_summary=pd.merge(tmp_df_summary,tmp_df_failed,left_index=True,right_index=True)
    tmp_df_summary["success_ratio"]= tmp_df_summary.success / tmp_df_summary.all_
    del tmp_df_all
    del tmp_df_success
    del tmp_df_failed
    return tmp_df_summary

```

## データのロードと加工

### データのロード

```python
# 元データCSVをロード
train_ksp=pd.read_csv("./in_data/ks-projects-201801.csv")

# canceled,live,suspended,undefinedのレコードを削除
train_ksp.drop(train_ksp[train_ksp.state.isin(["canceled","live","suspended","undefined"])].index,inplace=True)
```

### 開始月、開催期間の生成

```python
# 開始月 funding_start_month と 開催期間（週単位）funding_period_of_weeks を生成
train_ksp["funding_start_month"]    =pd.to_datetime(train_ksp.launched).dt.month
train_ksp["funding_period_of_weeks"]=(
    (
        pd.to_datetime(train_ksp.deadline)-pd.to_datetime(train_ksp.launched)
    ).dt.days/7
).astype(int)
```

### nameの文字数、単語数の生成

```python
# nameの文字数 name_char_count を生成
train_ksp["name_char_count"]        =train_ksp.name.str.strip().str.len()
train_ksp.name_char_count.fillna(0,inplace=True)
train_ksp["name_char_count"]        =train_ksp["name_char_count"].astype(int)

# nameの単語数 name_word_count を生成
def tmp_func_word_count(data):
    counter=0
    for aaa in re.split('[,./\s]',data.strip()):
        counter+=1
    return counter
tmp_universal_func_word_count=np.frompyfunc(tmp_func_word_count,1,1)
train_ksp["name_word_count"]=tmp_universal_func_word_count(train_ksp.name.astype(str))
train_ksp["name_word_count"].fillna(0,inplace=True)
train_ksp["name_word_count"]=train_ksp["name_word_count"].astype(int)
```

```python
# 使用しない項目の削除
train_ksp.drop(columns=["ID","goal","pledged","backers","usd pledged","usd_pledged_real","launched","deadline","name"],inplace=True)

display(train_ksp.info())
display(train_ksp.head())
```

## 量的変数の標準化

```python
#tmp_target_col_list=["usd_goal_real","funding_start_month","funding_period_of_weeks","name_char_count","name_word_count"]
tmp_target_col_list=["usd_goal_real","funding_period_of_weeks","name_char_count","name_word_count"]
tmp_target=train_ksp[tmp_target_col_list]

# 標準化用オブジェクトの作成
scaler = StandardScaler() 
# 標準化のデータ解析実施。標準化の対象列、並び順も含めて間違えないように注意すること。
scaler.fit(tmp_target.astype(float))
# 標準化のデータ編集実施。対象列、並び順も含めて間違えないように注意すること。
tmp_target_scaled=scaler.transform(tmp_target.astype(float))


print("---標準化前のデータ　サンプル表示---------------------")
display(train_ksp.head())

train_ksp_scaled=train_ksp.copy()
train_ksp_scaled[tmp_target_col_list]=tmp_target_scaled

print("---標準化後のデータ　サンプル表示---------------------")
display(train_ksp_scaled.head())

print("---標準化解析結果　StandardScaler解析後パラメータ(項目名リスト、平均値リスト、分散リスト)-----")
print(tmp_target_col_list)
print(scaler.mean_)
print(scaler.var_)


print("---標準化前の各項目の統計情報---------------------")
for tmp_aaa in tmp_target_col_list:
    print("{} mean={:.5f}, min={:.5f}, max={:.5f}, std={:.5f}".format(
        tmp_aaa,
        train_ksp[tmp_aaa].mean(),
        train_ksp[tmp_aaa].min(),
        train_ksp[tmp_aaa].max(),
        train_ksp[tmp_aaa].std(),
    ))

    
print("---標準化後の各項目の統計情報---------------------") 
for tmp_aaa in tmp_target_col_list:
    print("{} mean={:.5f}, min={:.5f}, max={:.5f}, std={:.5f}".format(
        tmp_aaa,
        train_ksp_scaled[tmp_aaa].mean(),
        train_ksp_scaled[tmp_aaa].min(),
        train_ksp_scaled[tmp_aaa].max(),
        train_ksp_scaled[tmp_aaa].std(),
    ))

del tmp_target_col_list,tmp_target,tmp_target_scaled,tmp_aaa

```

## 量的変数の無相関化

各項目の詳細分析では、量的変数の間に多重共線性は認められなかったので、無相関化する意味はなさそう。<br>
しかし、無相関化しても悪影響は無さそうなので、学習の意味も込めて、無相関化を導入する。<br>
なお、DAY2では暫定措置として、分散共分散行列の固有ベクトルを利用して無相関化するが、DAY4で主成分分析のライブラリを活用して、本格的に行う予定。

```python
# 無相関化の処理内容深堀りはDay4で行う。
# 今回は教わった処理をそのまま実行する。
start=time.time()

# cov = np.cov(train_X, rowvar=0) # 分散・共分散を求める
# _, S = np.linalg.eig(cov)           # 分散共分散行列の固有ベクトルを求める。
# train_X_decorr = np.dot(S.T,train_X).T #データを無相関化


#tmp_target_col_list=["usd_goal_real","funding_start_month","funding_period_of_weeks","name_char_count","name_word_count"]
tmp_target_col_list=["usd_goal_real","funding_period_of_weeks","name_char_count","name_word_count"]
tmp_target=train_ksp_scaled[tmp_target_col_list]


cov = np.cov(tmp_target, rowvar=0) # 分散・共分散を求める
_, S = np.linalg.eig(cov)           # 分散共分散行列の固有ベクトルを求める。
tmp_target_decorr = np.dot(S.T,tmp_target.T).T #データを無相関化


train_ksp_scaled_decorr=train_ksp_scaled.copy()
train_ksp_scaled_decorr[tmp_target_col_list]=tmp_target_decorr


print("---無相関化前のデータ　サンプル表示---------------------")
display(train_ksp_scaled.head())


print("---無相関化後のデータ　サンプル表示---------------------")
display(train_ksp_scaled_decorr.head())

print("---無相関化解析結果-------------------------------------")
print("---分散共分散行列-----")
print(tmp_target_col_list)
print(cov)
print("---固有値-----")
print(tmp_target_col_list)
print(_)
print("---固有ベクトル-----")
print(tmp_target_col_list)
print(S)


print("---無相関化前の各項目の相関関係--------------------------")
# 相関グラフ
sns.pairplot(train_ksp_scaled[tmp_target_col_list])
plt.show()
# 相関係数のテーブル表示
display(train_ksp_scaled[tmp_target_col_list].corr())

# 相関係数のヒートマップ表示
sns.heatmap(train_ksp_scaled[tmp_target_col_list].corr(),cmap="RdBu",vmin=-1,vmax=1)
plt.show()

print("---無相関化後の各項目の相関関係--------------------------")
# 相関グラフ
sns.pairplot(train_ksp_scaled_decorr[tmp_target_col_list])
plt.show()
# 相関係数のテーブル表示
display(train_ksp_scaled_decorr[tmp_target_col_list].corr())

# 相関係数のヒートマップ表示
sns.heatmap(train_ksp_scaled_decorr[tmp_target_col_list].corr(),cmap="RdBu",vmin=-1,vmax=1)
plt.show()




del tmp_target_col_list,tmp_target
del cov,_,S,tmp_target_decorr
del train_ksp_scaled

stop=time.time()
print("%.3f seconds" % (stop - start))
```

## モデル１の評価（ロジスティック回帰、ハイパーパラメータの手作業探索）

* ロジスティック回帰、elasticnet正則化ありのモデルを使用する。
* 最適なハイパーパラメータの探索を手作業で行う。（本格的な探索はグリッドサーチを用いて行う。本セクションは手作業探索を体感する目的）
* モデルの評価には、交差検証を用いる。データ全件を5分割して交差検証を行い、各イテレーションの評価指標を平均したものをモデルの性能とする。

```python
# ハイパーパラメータの設定値変更とセル実行を繰り返して、各設定値に対する評価指標を得ることで、手作業探索を行う。

# -------------------------------------------
# ハイパーパラメータの設定
alpha=0.01
l1_ratio=0.001



start=time.time()

# ----------------------------------------------------------
# 前処理（共通）後データのロード
train_X=pd.read_pickle("./out_data/train_X.pkl")
train_Y=pd.read_pickle("./out_data/train_Y.pkl")


# -----------------------------------
# 処理で使う変数の初期化
split_sum_accuracy  = 0
split_sum_precision = 0
split_sum_recall    = 0
split_sum_f1_score  = 0
split_counter = 0


# -----------------------------------
# 交差検証の繰り返し処理
# テスト役を交代させながら学習と評価を繰り返す。
# n_splitsはデータを分割するグループ数
for train_idx, test_idx in KFold(n_splits=5, shuffle=True, random_state=1234).split(train_X,train_Y):
    split_counter += 1

    print("------------------------------------")
    print("交差検証イテレーション No.{}".format(split_counter))

    # ---------------------------------------------------
    # 学習用、テスト用に分割
    # 学習用データ
    X_tr = train_X.iloc[train_idx,:]
    y_tr = train_Y.iloc[train_idx,:]
    # テスト用データ
    X_test = train_X.iloc[test_idx,:]
    y_test = train_Y.iloc[test_idx,:]

    # ---------------------------------------------------
    # 入力データ整形
    # 使用するモデルの入力データ要件に合わせて、データ整形する。
    # 　・説明変数データはデータ件数の分だけ行があり、説明変数の数だけ列がある２次元配列。shape出力は"(XXXXX,XXXXX)"
    # 　・目的変数データははデータ件数の分だけ行があり、列が１列の１次元配列。shape出力は"(XXXXX, )"
    # 　・説明変数データと目的変数データのデータ行数が一致している。
    #
    # SGDClassifierの場合
    # 目的変数は1次元に変換する。そうしないと警告が出る。
    y_tr = y_tr.values.ravel()     #(XXXXX,1)->(XXXXX,)  列が１列の１次元配列に整形する。
    y_test = y_test.values.ravel() #(XXXXX,1)->(XXXXX,)  列が１列の１次元配列に整形する。
    
    
    # ---------------------------------------------------
    # モデル構築と学習
    # ロジスティック回帰、正則化あり（elasticnet）
    clf=SGDClassifier(alpha=alpha,l1_ratio=l1_ratio,  loss="log",penalty="elasticnet",max_iter=10000,fit_intercept=True,random_state=1234,tol=1e-3)
    clf.fit(X_tr,y_tr)
    
    # ---------------------------------------------------
    # 予測
    # テスト用データの説明変数に対して、目的変数を予測する。
    y_est=clf.predict(X_test)
    # テスト用データの説明変数に対して、予測した目的変数値の確率を求める。
    y_est_proba=clf.predict_proba(X_test)
    
    # ---------------------------------------------------
    # 対数尤度を算出
    # log_lossは”負の”対数尤度を計算する関数であるため、マイナスを付けて本来の対数尤度に戻す。
    # normalizeオプションがTrueのままでは、合計の対数尤度ではなく平均の対数尤度が計算されるので注意。合計の対数尤度を求めたいのでFalseとする。
    log_likelihood = - log_loss(y_test, y_est_proba,normalize=False)
    print("対数尤度          = {:.3f}".format(log_likelihood))
    
    # ---------------------------------------------------
    # 評価指標を算出
    accuracy = accuracy_score(y_test, y_est)
    precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_est)
    print("正答率(Accuracy)  = {:.3f}%".format(100 * accuracy    ))
    print("適合率(Precision) = {:.3f}%".format(100 * precision[1]))
    print("再現率(Recall)    = {:.3f}%".format(100 * recall[1]   ))
    print("F1値(F1-score)    = {:.3f}%".format(100 * f1_score[1] ))

    # ---------------------------------------------------
    # 評価指標を加算
    split_sum_accuracy  +=accuracy
    split_sum_precision +=precision[1]
    split_sum_recall    +=recall[1]
    split_sum_f1_score  +=f1_score[1]


# 評価指標の平均化
final_accuracy  = split_sum_accuracy  / split_counter
final_precision = split_sum_precision / split_counter
final_recall    = split_sum_recall    / split_counter
final_f1_score  = split_sum_f1_score  / split_counter

print("\n\n")
print("====================================")
print("ハイパーパラメータ = alpha={},l1_ratio={}".format(alpha,l1_ratio))
print("交差検証結果（評価指標の平均値）")
print("　正答率(Accuracy)  = {:.3f}%".format(100 * accuracy    ))
print("　適合率(Precision) = {:.3f}%".format(100 * precision[1]))
print("　再現率(Recall)    = {:.3f}%".format(100 * recall[1]   ))
print("　F1値(F1-score)    = {:.3f}%".format(100 * f1_score[1] ))

stop=time.time()
print("%.3f seconds" % (stop - start))

del split_sum_accuracy, split_sum_precision, split_sum_recall, split_sum_f1_score, split_counter
del clf, y_tr, y_test, y_est, y_est_proba,log_likelihood 
del accuracy, precision, recall, f1_score
del final_accuracy, final_precision, final_recall, final_f1_score
del X_test, X_tr, alpha, l1_ratio, start, stop, test_idx, train_X, train_Y, train_idx
```

## モデル２の評価（ロジスティック回帰、ハイパーパラメータのグリッドサーチ探索）

* ロジスティック回帰を用いる。

```python
start=time.time()


# ----------------------------------------------------------
# 前処理（共通）後データのロード
train_X=pd.read_pickle("./out_data/train_X.pkl")
train_Y=pd.read_pickle("./out_data/train_Y.pkl")




# ----------------------------------------------------------
# モデル入力データの整形
#  train_X  #(331675,2442) 説明変数の数だけ次元がある。この形のまま扱えるので整形する必要はない。
train_Y_1d=train_Y.values.ravel() #(331675,1)ｰ>(331675,)　２次元になっている。１次元にしないと警告が出る。


# ----------------------------------------------------------
# データ全件からランダムで抽出した10万件で交差検証して、3000件でテストする。
# まず103000件をランダムに抽出（全331675件✕0.31054496）して、それを10万件と3000件にランダムに分割（103000件✕0.02912621）
# _1,_2は使わないデータ
_1, X_tmp, _2, y_tmp= train_test_split(train_X, train_Y_1d, test_size=0.31054496)
X_tr, X_test, y_tr, y_test= train_test_split(X_tmp, y_tmp, test_size=0.02912621)

del _1,_2








# ----------------------------------------------------------
# モデル構築
# グリッドサーチで可変にしたいパラメータは、インスタンス作成の段階では指定しない。
model=SGDClassifier(loss="log",penalty="elasticnet",max_iter=10000,fit_intercept=True,random_state=1234,tol=1e-3)


# ----------------------------------------------------------
# グリッドサーチするハイパーパラメータセット
parameters=[
#    {'alpha': [0.01, 0.1], 'l1_ratio': [0.001,0.5,1] }
    {'alpha': [0.000001,0.00001,0.0001,0.001,0.01, 0.1,1,10,100], 'l1_ratio': [0.0001, 0.001, 0.01, 0.1, 1] }
]


# ----------------------------------------------------------
# グリッドサーチを構築する。
# cvは交差検証の分割グループ数、scoringはベストを判定する評価指標、n_jobsは並列数（-1は自動判定）
# verboseは進捗ログを出力するペースをイテレーション数で指定する。（2の場合、イテレーション２回ごとに１行ログを出力する）
clf = GridSearchCV(model,parameters, cv=5,n_jobs=-1,verbose=1)

# 学習する。
# 交差検証のグループ分けと反復処理は、グリッドサーチのfit関数の中で実施される。
clf.fit(X_tr,y_tr)


# グリッドサーチが判定した最適なハイパーパラメータセットを出力する。
print(clf.best_params_, clf.best_score_)


# ----------------------------------------------------------
# グリッドサーチで決定した最適なハイパーパラメータを用いてモデルを構築して、テストを行う。

# まず、データ全件（331675）を用いて最適なハイパーパラメータセットで学習して、
# テスト用データ3000件に対して予測して、各種評価指標を得る。
best_model =SGDClassifier(**clf.best_params_,loss="log",penalty="elasticnet",max_iter=10000,fit_intercept=True,random_state=1234,tol=1e-3)
best_model.fit(train_X, train_Y_1d)
print("識別精度=",best_model.score(X_test, y_test)) #accuracyのスコアを表示している。


# テスト用データの説明変数に対して、目的変数を予測する。
y_est=best_model.predict(X_test)
# テスト用データの説明変数に対して、目的変数の確率値を予測する。
y_est_proba=best_model.predict_proba(X_test)


# 対数尤度を算出
# log_lossは”負の”対数尤度を計算する関数であるため、マイナスを付けて本来の対数尤度に戻す。
# normalizeオプションがTrueのままでは、合計の対数尤度ではなく平均の対数尤度が計算されるので注意。合計の対数尤度を求めたいのでFalseとする。
log_likelihood = - log_loss(y_test, y_est_proba,normalize=False)
print("対数尤度（log_loss関数出力）：{:.3f}".format(log_likelihood))


# 重みを取得
print("切片の数：",best_model.intercept_.shape)
print("重み係数の数：",best_model.coef_.shape)
print("---切片の値-------------")
print(best_model.intercept_)
print("---重み係数の値---------")
print(best_model.coef_)


# ライブラリを使用して評価指標を算出
accuracy = accuracy_score(y_test, y_est)
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_est)

# 評価指標の表示
print("---ライブラリを使用して算出した評価指標-----------")
print("正答率(Accuracy)  = {:.3f}%".format(100 * accuracy    ))
print("適合率(Precision) = {:.3f}%".format(100 * precision[1]))
print("再現率(Recall)    = {:.3f}%".format(100 * recall[1]   ))
print("F1値(F1-score)    = {:.3f}%".format(100 * f1_score[1] ))


# 混同行列を表示
display(
    pd.DataFrame(
        confusion_matrix(y_test,y_est),
        index=["真（failed）","真（successful）"],
        columns=["推定（failed）","推定（successful）"])
       )

stop=time.time()
print("%.3f seconds" % (stop - start))

del train_Y_1d, X_tr, X_test, y_tr, y_test,model, parameters
del clf, best_model
del y_est, y_est_proba, log_likelihood, accuracy, precision, recall, f1_score, _
del start, stop
del train_X, train_Y
```

### 使用単語リスト暫定版の生成

```python
# %%timeは並列処理でうまく動作しないので、time()を使う。
start = time.time()

# ---調査対象データの準備--------------------------------------------------------------------------------------
# 元データをインポート
tmp_name=pd.read_csv("./in_data/ks-projects-201801.csv")
# canceled,live,suspended,undefinedの案件を削除
tmp_name.drop(tmp_name[tmp_name.state.isin(["canceled","live","suspended","undefined"])].index,inplace=True)
# name だけ使用する。
tmp_name=tmp_name.loc[:,["name","state"]]


# ---使用単語リストの作成--------------------------------------------------------------------------------------
# 処理時間短縮のため、分割して並列処理する。各ワーカープロセスに任せるデータ件数を設定。
batch_size= 50000

# バッチ数の算出
workers_size=math.ceil(tmp_name.shape[0] / batch_size)

print("データ件数=",tmp_name.shape[0])
print("バッチサイズ=",batch_size)
print("バッチ数=",workers_size)

# データを各ワーカープロセスに振り分けるためにバッチ分けしてリスト化
tmp_name_batch_list=[]
for i in range(1,workers_size+1):
    from_idx =batch_size * (i-1)
    to_idx   =batch_size * i if tmp_name.shape[0] > (batch_size * i) else tmp_name.shape[0]
    tmp=tmp_name[from_idx:to_idx]
    print("バッチNo.={:},\t行番号範囲={:} - {:},\t件数={:}".format(i,tmp.head(1).index[0],tmp.tail(1).index[0],tmp.shape[0]))
    tmp_name_batch_list.append(tmp)

    
# 並列処理内容を定義
# 各ワーカープロセスでは、担当したデータの各案件名nameで使われている単語を調べた結果をデータフレームにして戻す。
def parallel_task_word_count(df):
    # 単語をインデックスにした、成功件数と失敗件数の２列のデータフレーム作成。wcはword countの略
    tmp_wc_rank_list=pd.DataFrame(index=[],columns=["successful","failed"])
    # データの各行で、nameを単語に分解して成功失敗を計数する。
    for row in df.itertuples():
        if (row.name is None) | (type(row.name) is float):
            continue
        
        # 区切り記号で単語に分割して、単語毎に繰り返す。
        # -と_は単語が連結した１つの単語とみなすと考えて、単語分割の区切り記号には入れない。例）state-of-the-art、state_of_the_art
        
        # １案件内で単語が何回使用されているかは無視して、１回以上使用されていれば頻度１としたい。
        # 一旦セットに登録することで、それを実現する。
        words_set=set()
        
        for www in (aaa.lower() for aaa in re.split("[<>~#®•=—@\*\?\|\'\"\(\)\{\}\[\]\_\-\–\+;:/,.!\s]",row.name.strip()) if aaa !=''):
            words_set.add(www)
        
        # セットに登録された単語毎に、ランキング表に追加する。
        for www in words_set:
            # その単語の行がランキング表に存在しなければ、行を追加する。
            if www not in tmp_wc_rank_list.index.values:
                tmp_wc_rank_list.loc[www]=[0,0]
            # 既存行に+1する。
            tmp_wc_rank_list.loc[www,row.state]+=1
        
    return tmp_wc_rank_list


# ワーカープロセスを並列で生成して、並列処理させる。
# バッチ分けしたリストを引数にすることで、リストの各要素（バッチ）を各ワーカーに引き渡す。
# 各ワーカープロセスからの戻り値である使用頻度ランキング表を回収する。
with concurrent.futures.ProcessPoolExecutor(max_workers=workers_size) as executor:
    result_list=list(executor.map(parallel_task_word_count, tmp_name_batch_list))

    
# 各ワーカープロセスが生成した使用頻度ランキング表を統合して、全体としての使用頻度ランキング表を生成する。
integrated_wc_rank_list=pd.DataFrame(index=[],columns=["successful","failed"])
tmp_count=0
for aaa in result_list:
    tmp_count += 1
    print("---使用頻度ランキング表の統合 {}/{}.---".format(tmp_count,len(result_list)))
    for row in aaa.itertuples():
        if row.Index not in integrated_wc_rank_list.index.values:
            integrated_wc_rank_list.loc[row.Index]=[0,0]
        integrated_wc_rank_list.loc[row.Index,"successful"]+=row.successful
        integrated_wc_rank_list.loc[row.Index,"failed"]+=row.failed

# 統合版ランキング表で、all_、success_ratioの２列を追加（属性allと競合するので_を付けている）
integrated_wc_rank_list["all_"]          =integrated_wc_rank_list.successful + integrated_wc_rank_list.failed
integrated_wc_rank_list["success_ratio"] =integrated_wc_rank_list.successful / integrated_wc_rank_list.all_
# 統合版ランキング表を、単語の使用頻度で降順ソート。第２キーは表の画面出力の視認性を上げるのが目的。
# 本当は、使用頻度all_と単語名（インデックス）の複合キーでソートしたいが、やり方がわからないので保留。
integrated_wc_rank_list.sort_values(["all_","success_ratio"], ascending=[False,False],inplace=True)

# display(integrated_wc_rank_list)

stop = time.time()
print('%.3f seconds' % (stop - start))

del batch_size,workers_size,tmp_name_batch_list,executor,result_list
del aaa, i, row, from_idx, to_idx,tmp,tmp_count
del start, stop
```

### name列での使用単語をone-hot化

```python
start=time.time()

# ---調査対象データの準備--------------------------------------------------------------------------------------
# 元データをインポート
tmp_name=pd.read_csv("./in_data/ks-projects-201801.csv")
# canceled,live,suspended,undefinedの案件を削除
tmp_name.drop(tmp_name[tmp_name.state.isin(["canceled","live","suspended","undefined"])].index,inplace=True)
# name だけ使用する。
tmp_name=tmp_name.loc[:,["name","state"]]




# 対象単語リストは、リスト生成時点で閾値以上の使用頻度に限定した単語リストになっている。
# リストにある単語について、one-hot化する。



# ダミー変数列を追加
for word in one_hot_target_word_list:
    tmp_name["name_wd_"+word]=0


# CPU数に合わせてバッチ分けするための、バッチサイズとバッチ数（ワーカー数）算出
import os
num_process=os.cpu_count()
batch_size= int(tmp_name.shape[0]/num_process)
workers_num=math.ceil(tmp_name.shape[0]/batch_size)

print("データ件数=",tmp_name.shape[0])
print("バッチサイズ=",batch_size)
print("バッチ数=",workers_num)

# データをワーカープロセスに振り分けるためにバッチ分けしてリスト化
batch_list = [tmp_name[i:((i+batch_size) if i+batch_size<tmp_name.shape[0] else tmp_name.shape[0])] for i in range(0, tmp_name.shape[0], batch_size)]



# 並列処理内容を定義
def parallel_task_word_one_hot_encode(df):
    # nameを単語に分解して、対象単語リストに存在する場合は該当列（ダミー変数）に1を立てる。
    for idx,row in df.iterrows(): # itertuples()では、カラム数が多くなるとnamedtupleではなくtupleになってしまう!!!!!!!!!
        if (row["name"] is None) | (type(row["name"]) is float):
            continue
        # nameで使用されている単語の種類を抽出。一旦セットに格納することで、種類数としてまとめる。
        words_set=set()
        for www in (aaa.lower() for aaa in re.split("[<>~#®•=—@\*\?\|\'\"\(\)\{\}\[\]\_\-\–\+;:/,.!\s]",row["name"].strip()) if aaa !=''):
            words_set.add(www)
        
        # セットに登録された単語毎に、対象単語リストと突合。ヒットしたらダミー変数に1を立てる。
        for www in words_set:
            if www in one_hot_target_word_list:
                df.loc[idx,"name_wd_"+www]=1
    return df


# ワーカープロセスを生成して、並列処理させる。
# バッチ分けしたリストを引数にすることで、リストの各要素（バッチ）を各ワーカーに引き渡す。
# 各ワーカープロセスの処理結果は、引数として与えたデータのバッチに反映されている。
with concurrent.futures.ProcessPoolExecutor(max_workers=workers_num) as executor:
    result_list=list(executor.map(parallel_task_word_one_hot_encode, batch_list))

# 処理結果が反映された各バッチを統合して、処理結果データフレームを生成する。
tmp_name=pd.concat(result_list)

stop=time.time()
print("%.3f seconds" % (stop - start))

del num_process,batch_size,workers_num,batch_list
del executor,one_hot_target_word_list
```

```python
start=time.time()


# ----------------------------------------------------------
# モデル入力データの整形
#  train_X  #(331675,2442) 説明変数の数だけ次元がある。この形のまま扱えるので整形する必要はない。
train_Y_1d=train_Y.values.ravel() #(331675,1)ｰ>(331675,)　２次元になっている。１次元にしないと警告が出る。


# ----------------------------------------------------------
# 処理が数時間経っても終わらないので、データ件数を減らす。
# データ全件からランダムで抽出した10万件で交差検証して、30000件でテストする。
# まず103000件をランダムに抽出（全331675件✕0.31054496）して、それを10万件と3000件にランダムに分割（103000件✕0.02912621）
# _1,_2は使わないデータ
_1, X_tmp, _2, y_tmp= train_test_split(train_X, train_Y_1d, test_size=0.31054496)
X_tr, X_test, y_tr, y_test= train_test_split(X_tmp, y_tmp, test_size=0.02912621)

del _1,_2





# ----------------------------------------------------------
# モデル構築
# グリッドサーチで可変にしたいパラメータは、インスタンス作成の段階では指定しない。
model=SGDClassifier(loss="log",penalty="elasticnet",max_iter=10000,fit_intercept=True,random_state=1234,tol=1e-3)


# ----------------------------------------------------------
# グリッドサーチするハイパーパラメータセット
parameters=[
#    {'alpha': [0.01, 0.1], 'l1_ratio': [0.001,0.5,1] }
    {'alpha': [0.000001,0.00001,0.0001,0.001,0.01, 0.1,1,10,100], 'l1_ratio': [0.0001, 0.001, 0.01, 0.1, 1] }
]


# ----------------------------------------------------------
# グリッドサーチを構築する。
# cvは交差検証の分割グループ数、scoringはベストを判定する評価指標、n_jobsは並列数（-1は自動判定）
# verboseは進捗ログを出力するペースをイテレーション数で指定する。（2の場合、イテレーション２回ごとに１行ログを出力する）
clf = GridSearchCV(model,parameters, cv=5,n_jobs=-1,verbose=1)

# 学習する。
# 交差検証のグループ分けと反復処理は、グリッドサーチのfit関数の中で実施される。
clf.fit(X_tr,y_tr)


# グリッドサーチが判定した最適なハイパーパラメータセットを出力する。
print(clf.best_params_, clf.best_score_)


# ----------------------------------------------------------
# グリッドサーチで決定した最適なハイパーパラメータを用いてモデルを構築して、テストを行う。

# まず、データ全件（331675）を用いて最適なハイパーパラメータセットで学習して、
# テスト用データ3000件に対して予測して、各種評価指標を得る。
best_model =SGDClassifier(**clf.best_params_,loss="log",penalty="elasticnet",max_iter=10000,fit_intercept=True,random_state=1234,tol=1e-3)
best_model.fit(train_X, train_Y_1d)
print("識別精度=",best_model.score(X_test, y_test)) #accuracyのスコアを表示している。


# テスト用データの説明変数に対して、目的変数を予測する。
y_est=best_model.predict(X_test)
# テスト用データの説明変数に対して、目的変数の確率値を予測する。
y_est_proba=best_model.predict_proba(X_test)


# 対数尤度を算出
# log_lossは”負の”対数尤度を計算する関数であるため、マイナスを付けて本来の対数尤度に戻す。
# normalizeオプションがTrueのままでは、合計の対数尤度ではなく平均の対数尤度が計算されるので注意。合計の対数尤度を求めたいのでFalseとする。
log_likelihood = - log_loss(y_test, y_est_proba,normalize=False)
print("対数尤度（log_loss関数出力）：{:.3f}".format(log_likelihood))


# 重みを取得
print("切片の数：",best_model.intercept_.shape)
print("重み係数の数：",best_model.coef_.shape)
print("---切片の値-------------")
print(best_model.intercept_)
print("---重み係数の値---------")
print(best_model.coef_)


# ライブラリを使用して評価指標を算出
accuracy = accuracy_score(y_test, y_est)
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_est)

# 評価指標の表示
print("---ライブラリを使用して算出した評価指標-----------")
print("正答率(Accuracy)  = {:.3f}%".format(100 * accuracy    ))
print("適合率(Precision) = {:.3f}%".format(100 * precision[1]))
print("再現率(Recall)    = {:.3f}%".format(100 * recall[1]   ))
print("F1値(F1-score)    = {:.3f}%".format(100 * f1_score[1] ))


# 混同行列を表示
display(
    pd.DataFrame(
        confusion_matrix(y_test,y_est),
        index=["真（failed）","真（successful）"],
        columns=["推定（failed）","推定（successful）"])
       )

stop=time.time()
print("%.3f seconds" % (stop - start))

del train_Y_1d, X_tr, X_test, y_tr, y_test,model, parameters
del clf, best_model
del y_est, y_est_proba, log_likelihood, accuracy, precision, recall, f1_score, _
del start, stop
```

## モデル７の評価（サポートベクターマシン）

サポートベクターマシンを用いる。<br>
最適なハイパーパラメータの探索をグリッドサーチで行う。<br>
モデルの評価には、交差検証を用いる。<br>

```python
start=time.time()

# ----------------------------------------------------------
# 前処理（共通）後データのロード
train_X=pd.read_pickle("./out_data/train_X.pkl")
train_Y=pd.read_pickle("./out_data/train_Y.pkl")


# ----------------------------------------------------------
# 目的変数0,1の２値を-1,1の２値に変更する。
train_Y.state_successful=train_Y.state_successful.replace(0,-1)


# ----------------------------------------------------------
# モデル入力データの整形
#  train_X_svc  #(331675,2442) 説明変数の数だけ次元がある。この形のまま扱えるので整形する必要はない。
train_Y_1d=train_Y.values.ravel() #(331675,1)ｰ>(331675,)　２次元になっている。１次元にしないと警告が出る。


# ----------------------------------------------------------
# 処理が数時間経っても終わらないので、データ件数を減らす。
# データ全件からランダムで抽出した1万件で交差検証して、30000件でテストする。
# まず13000件をランダムに抽出（全331675件✕0.039195）して、それを1万件と3000件にランダムに分割（13000件✕0.2307692）
# _1,_2は使わないデータ
_1, X_tmp, _2, y_tmp= train_test_split(train_X, train_Y_1d, test_size=0.039195)
X_tr, X_test, y_tr, y_test= train_test_split(X_tmp, y_tmp, test_size=0.2307692)

del _1,_2




# ----------------------------------------------------------
# モデル構築
# グリッドサーチで可変にしたいパラメータは、この段階では指定しない。
model=SVC()

# ----------------------------------------------------------
# グリッドサーチするパラメータセット（サポートベクターマシンSVCの場合）
parameters=[
#    {'C': [1, 5, 10, 100, 1000], 'kernel': ['rbf'], 'gamma': ["scale"] },
    {'C': [0.01], 'kernel': ['rbf'], 'gamma': ["scale"] },
#    {'C': [1, 5, 10, 100, 1000], 'kernel': ['linear'] },
#    {'C': [1, 5, 10, 100, 1000], 'kernel': ['poly'], 'degree': [2, 3, 4], 'gamma': ["scale"] },
#    {'C': [1, 5, 10, 100, 1000], 'kernel': ['sigmoid'], 'gamma': ["scale"] }
]
# C コストパラメータ
#　　誤分類の許容度合い。
#    値が大きい方が厳密に分類する（識別平面の傾きが急。誤分類を許容せず、厳密に複雑に境界する）。
#    値が小さい方が大局的に分類する。（識別平面の傾きが緩い。誤分類を許容して、シンプルに境界する。）
# gamma
#　　訓練データの各要素が距離的にどこまで遠くの予測対象要素に影響を及ぼすか。
#    カーネル関数rbf（ガウスカーネル）を選択するときはgammaの値も必要。gmma='scale'にすると自動計算される
# cv
#　　交差検証のグループ数
# n_jobs
#　　並列処理数




# ----------------------------------------------------------
# グリッドサーチを構築、学習して最適なパラメータセットを評価する。
#
# 交差検証のグループ数を設定
n_split = 5 
# グリッドサーチを構築する
# cvは交差検証の分割グループ数、scoringはベストを判定する評価指標、n_jobsは並列数（-1は自動判定）
clf = GridSearchCV(model, parameters, cv=5, n_jobs=-1,scoring="f1",verbose=1) 

# 学習する。
# 交差検証のグループ分けと反復処理は、グリッドサーチのfit関数の中で実施される。
clf.fit(X_tr,y_tr)


# グリッドサーチが判定した最適なパラメータセットを出力する。
print(clf.best_params_, clf.best_score_)


# ----------------------------------------------------------
# 訓練データ全件に対して最適なパラメータセットで学習して、
# 学習後モデルと各種評価指標を得る。
best_model = SVC(**clf.best_params_,probability=True) 
best_model.fit(X_tr, y_tr)
print("識別精度=",best_model.score(X_test, y_test)) #accuracyのスコアを表示している。


# 最終テスト用データの説明変数に対して、目的変数を予測する。
y_est=best_model.predict(X_test)


# ライブラリを使用して評価指標を算出
accuracy = accuracy_score(y_test, y_est)
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_est)

# 評価指標の表示
print("---ライブラリを使用して算出した評価指標-----------")
print("正答率(Accuracy)  = {:.3f}%".format(100 * accuracy    ))
print("適合率(Precision) = {:.3f}%".format(100 * precision[1]))
print("再現率(Recall)    = {:.3f}%".format(100 * recall[1]   ))
print("F1値(F1-score)    = {:.3f}%".format(100 * f1_score[1] ))


# 混同行列を表示
display(
    pd.DataFrame(
        confusion_matrix(y_test,y_est),
        index=["真（failed）","真（successful）"],
        columns=["推定（failed）","推定（successful）"])
)

stop=time.time()
print("%.3f seconds" % (stop - start))

del train_Y_1d, X_tr, X_test, y_tr, y_test,model, parameters, n_split
del clf, best_model
del y_est,  accuracy, precision, recall, f1_score, _
del start, stop
```
